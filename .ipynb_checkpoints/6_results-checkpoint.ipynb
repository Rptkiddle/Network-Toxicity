# Analysis & Visualization

*This workbook repacks the results objects from the IDTxl estimation process (4_estimate.py) into arrays and produces several visualizations describing network toxicity.*

- **INPUTS:** the following input files must be in the data directory:
    - resampled.csv (resampled toxicity data, output of 4_resample.py)
    - results_dayx_to_dayy.pkl (mTE estimation results, output of 5_estimate.py)
    - global_IDs.pkl (global edge identifiers, output of 5_estimate.py)
    - convert_ID_dayx_to_dayy.pkl (local to global edge conversions, output of 5_estimate.py)
    - ch_stats.csv (channel statistics, output of 3_describe.ipynb)
    - ch_info.csv (channel information, manually specified file with categories from Simon et al 2022)

## setup

```python
#GENERAL
import pickle
import random
import itertools
import numpy as np
import pandas as pd
from scipy import stats

#GRAPHING
import igraph
import cairo
import pathpy as pp
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
from backbone_network import get_graph_backbone

#data directory:
datadir = '/Users/rptkiddle/Desktop/Network-Toxicity/data/'
```

## prepare data

```python
#load resampled data:
#resampled = pd.read_csv(datadir+'resampled.csv')
resampled = pd.read_csv('resampled.csv')

#reformat date column (str -> datetime): 
resampled['date']= pd.to_datetime(resampled['date'])

#set datetimeindex (time series data):
resampled = resampled.set_index('date').sort_index()

#pivot long into wide (sparse) format:
pivoted = pd.pivot(resampled, columns='source', values='retox')
```

```python
#load mTE results:
with open(datadir+'results_day0_to_day165.pkl', 'rb') as f:
    results = pickle.load(f)
```

```python
#load global IDs:
with open(datadir+'global_IDs.pkl', 'rb') as f:
    global_IDs = pickle.load(f)
    
#load ID conversion dict:
with open(datadir+'convert_ID_day0_to_day165.pkl', 'rb') as f:
    convert_ID = pickle.load(f)
```

```python
#load channel statistics (pre-resample):
ch_stats = pd.read_csv(datadir+'ch_stats.csv').drop(columns=['Unnamed: 0'])

#load channel information:
ch_info = pd.read_csv(datadir+'ch_info.csv')

#merge into channel reference dataframe:
ch_ref = pd.merge(left=ch_info, right=ch_stats, on='source', how='inner').sort_values(by='source').reset_index(drop=True)
```

## repack IDTxl results objects into arrays for plotting:

```python
# create template zeroed 3D array to populate with results:
x_sources = len(global_IDs)
y_targets = len(global_IDs)
z_windows = len(results)
template = np.zeros((x_sources, y_targets, z_windows))
print(template.shape)

#settings for sliding window (same as 4_estimate.py):
start_day = 0 #start here.
stop_day = 165 #stop here.
window_size = 96 #samples per window.
day_size = 288 #samples per day.
```

```python
# create 3D array with total counts of significant edges per time point:
sig_edges_xyz = np.copy(template)

for r, z in zip(results, range(sig_edges_xyz.shape[-1])):

    for targets in r[1].get_source_variables(fdr=False):
        target = targets['target']
        
        for sourceset in targets['selected_vars_sources']:
            edge = (convert_ID[r[0]][sourceset[0]], 
                    convert_ID[r[0]][target])
            sig_edges_xyz[...,z][edge[0],edge[1]] += 1
                       
#simplify to 2D array of total edge counts over all time (-z):
sig_edges_xy = np.sum(sig_edges_xyz, axis=2)
```

```python
# create arrays with total count of potential edges (all those considered).
tot_edges_xyz = np.copy(template)

begin, end  = start_day*window_size*int(day_size/window_size), start_day*window_size*int(day_size/window_size)+window_size

for r, z in zip(range(start_day*int(day_size/window_size), stop_day*int(day_size/window_size)), range(tot_edges_xyz.shape[-1])):
    #take window; drop any chats/channels for which there is missing data (nan):
    embedding = pivoted.iloc[begin:end].dropna(axis=1, how='any')
    sources = range(embedding.shape[1])
    #build list of tuples where each pair is the global coordinates (x,y) for a considered edge:
    local_coords = [c for c in list(itertools.product(sources, sources)) if len(set(c)) == len(c)]
    global_coords = [(convert_ID[str(embedding.first_valid_index())][x] , convert_ID[str(embedding.first_valid_index())][y]) for x,y in local_coords]
    #update array:
    for edge in global_coords:
        tot_edges_xyz[...,z][edge[0],edge[1]] += 1
    #increment window:
    begin += window_size
    end += window_size 
    
#simplify to 2D array (-z):
tot_edges_xy = np.sum(tot_edges_xyz, axis=2)
```

```python
# OPTION 1: overlay to create third array with proportion of significant edges / total edges.
#sigtot_xy = np.divide(sig_edges_xy, tot_edges_xy, out=np.zeros_like(sig_edges_xy), where=tot_edges_xy!=0)

#OPTION 2: overlay to create third array with proportion of significant edges / MAX total edges.
sigtot_xy = np.divide(sig_edges_xy, np.amax(tot_edges_xy), out=np.zeros_like(sig_edges_xy), where=tot_edges_xy!=0)
```

## plot toxicity transfer networks:

### filters (optional):

```python
#FILTER 1: minimum edge observation count (set to zero if using global sigtot measure):
f1_threshold = np.quantile(tot_edges_xy, q=0)
print(f1_threshold)

#FILTER 2: minimum sig/tot value (set to zero if using disparity filter below!):
f2_threshold = np.quantile(sigtot_xy, q=0)
print(f2_threshold)

#apply filters:
sigtot_xy_filter = np.copy(sigtot_xy)
sigtot_xy_filter[tot_edges_xy < f1_threshold] = 0
sigtot_xy_filter[sigtot_xy_filter <= f2_threshold] = 0
```

### build igraph object:

```python
#specify data:
data = sigtot_xy_filter

# get non-zero array indices:
indices = np.nonzero(data)
# ...generate edge list:
edges = zip(*indices)
# ...retrieve weights:
weights = data[indices]

#initialize directed graph object
G = igraph.Graph(edges=edges, directed=True)

#assign vertex attributes:
G.vs['NL'] = [x[0] for x in global_IDs.items()] #Dutch Chat/Channel Names
G.vs['EN'] = ch_info.sort_values(by='source').reset_index(drop=True)['translation'] #English Chat/Channel Names
G.vs['cats'] = ch_info.sort_values(by='source').reset_index(drop=True)['category'] #English Chat/Channel Categories
G.vs['type'] = ch_info.sort_values(by='source').reset_index(drop=True)['chat_channel'] #English Chat/Channel Type
G.vs['mtox'] = ch_ref['mean'] #mean toxicity levels.
G.vs['ttox'] = ch_ref['ttox'] #threshold toxicity levels.

#assign edge attributes:
G.es['weight'] = weights #edge weights
G.es['width'] = weights #edge widths
#G.es['width'] = [50*x for x in weights] #edge widths

#extract backbone:
BB = igraph.Graph.from_networkx(get_graph_backbone(G.to_networkx(), alpha_t=0.05))

#prune vertices with no edges:
BB.vs.select(_degree=0).delete()

#generate index for remaining edges:
BB.vs['ind'] = list(range(len(BB.vs['NL'])))

#create color list from mean tox values:
tox_rank = [int(x) for x in stats.rankdata(BB.vs['ttox'])]
num_colors = len(tox_rank)
palette = igraph.GradientPalette('yellow', 'red', num_colors+1)
color_list_tox= [palette.get(x) for x in tox_rank]
print(len(color_list_tox))
```

### plot network:

```python
#plot backbone w/ communities:
random.seed(1234)
igraph.plot(BB, mark_groups=BB.community_infomap(),
            vertex_color= color_list_tox, 
            vertex_label_color='white', vertex_frame_color='white',
            vertex_shape= ['square' if ch_type=='channel' else 'circle' for ch_type in BB.vs['type']],
            vertex_size=[600*x+5 for x in BB.pagerank()], vertex_label_dist=1, vertex_label_size=18,
            edge_color='black', edge_arrow_size= 0.5, edge_arrow_width=1,
            background='white',
            layout=BB.layout_fruchterman_reingold(niter=5000), bbox=(1000,600),
            #layout= BB.layout_davidson_harel(), bbox=(1600,1125))
            #layout= BB.layout_graphopt(), bbox=(1600,1125),
            target=datadir+'network.pdf')
```

### plot network w/ categories:

```python
#plot backbone w/ communities:
random.seed(1234)
igraph.plot(BB, mark_groups=BB.community_infomap(),
            vertex_color= 'blue', 
            vertex_label=BB.vs['cats'], vertex_label_color='white', vertex_frame_color='black',
            vertex_shape= ['square' if ch_type=='channel' else 'circle' for ch_type in BB.vs['type']],
            vertex_size=[600*x+5 for x in BB.pagerank()], vertex_label_dist=1, vertex_label_size=12,
            edge_color='black', edge_arrow_size= 0.2, edge_arrow_width=1,
            background='grey',
            layout=BB.layout_fruchterman_reingold(niter=5000), bbox=(1000,600),
            #layout= BB.layout_davidson_harel(), bbox=(1600,1125))
            #layout= BB.layout_graphopt(), bbox=(1600,1125),
            target=datadir+'categories.pdf')
```

### plot network w/ IDs and accompanying table:

```python
#save table with data for all IDs
ch_ID = BB.vs['ind']
ch_NL = BB.vs['NL']
ch_type = BB.vs['type']
ch_cat = BB.vs['cats']
ch_inf = BB.pagerank()
ttox = list(BB.vs['ttox'])
data = pd.DataFrame({'ID':ch_ID, 'name':ch_NL, 'type':ch_type, 'category':ch_cat, 'influence':ch_inf, 'toxicity':ttox})
data.to_latex(datadir+'text.tex', index=False)

#plot backbone w/ communities:
random.seed(1234)
igraph.plot(BB, mark_groups=BB.community_infomap(),
            vertex_label=BB.vs['ind'], vertex_color= 'purple', 
            vertex_label_color='white', vertex_frame_color='purple',
            vertex_shape= ['square' if ch_type=='channel' else 'circle' for ch_type in BB.vs['type']],
            vertex_size=[600*x+5 for x in BB.pagerank()], vertex_label_dist=1, vertex_label_size=18,
            edge_color='black', edge_arrow_size= 0.5, edge_arrow_width=1,
            background='grey', lty=2,
            layout=BB.layout_fruchterman_reingold(niter=5000), bbox=(1000,600),
            #layout= BB.layout_davidson_harel(), bbox=(1600,1125))
            #layout= BB.layout_graphopt(), bbox=(1600,1125),
            target=datadir+'networkID.pdf')
```

### calculate patterns of association:

```python
# a low-density network.
print(f'density of {BB.density()}')

#dissaortative network: nodes with high degree are more likely to connect to those with a low degree:
print(f'degree assortativity of {BB.assortativity_degree()}')

#nodes with a high level of toxicity are slightly more likely to connect with those of a lower level of toxicity:
print('mtox:')
print(BB.assortativity(BB.vs['mtox']))

#nodes with a high level of local influence are slightly more likely to connect with those with a lower level of influence.
print('local:')
print(BB.assortativity(BB.strength(mode='out')))

#nodes with a high level of global influence are slightly more likely to connect with those with a lower level of influence.
print('global:')
print(BB.assortativity(BB.pagerank()))

#nodes of the same type are considerably more likely to connect than those of a different category:
print('type:')
type_nom = pd.Series(BB.vs['type'],dtype="category").cat.codes
print(BB.assortativity_nominal(type_nom))

#nodes of the same category are slightly more likely to connect than those of a different category (but still a lot of mixing; see FYP): 
print('category:')
cat_nom = pd.Series(BB.vs['cats'],dtype="category").cat.codes
print(BB.assortativity_nominal(cat_nom))
```

### plot inward vs outward influence:

```python
#gather data for plot
ch_ID = BB.vs['EN']
ch_type = BB.vs['type']
str_out = BB.strength(mode='out', weights=BB.es['weight'])
str_in = BB.strength(mode='in', weights=BB.es['weight'])
mtox = list(BB.vs['mtox'])
ttox = list(BB.vs['ttox'])

#place in dataframe:
data = pd.DataFrame({'ch_ID':ch_ID, 'ch_type':ch_type, 'outward toxicity transfer':str_out, 'inward toxicity transfer':str_in, 'thres toxicity':mtox})

#build graph:
sns.set_theme(style='whitegrid')
sns.set(style='ticks')
sns.set(font_scale = 1.5)

g = sns.relplot(data=data, x='outward toxicity transfer', y='inward toxicity transfer',  hue='thres toxicity', size='thres toxicity',
    palette=sns.light_palette("red", as_cmap=True), sizes=(100,2000), height=8, aspect=1.5, legend=True)


for x, y, text in zip(str_out, str_in, ch_type):
    if x >=.15 or y>=.2: 
        g.axes[0,0].text(x,y+random.choice([-.01, 0, .01]),text, horizontalalignment='left')

#g.set(xscale="log", yscale="log")
g.ax.xaxis.grid(True, "minor", linewidth=.25)
g.ax.yaxis.grid(True, "minor", linewidth=.25)
g.despine(left=True, bottom=True)
plt.grid()
plt.savefig(datadir+'influencers.png')
plt.show()
```
